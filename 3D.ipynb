```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create a coordinate grid
# Define the range and resolution for the X and Y axes
x = np.linspace(-5, 5, 400) # 400 points provide a smooth appearance
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y) # Create a 2D grid from 1D arrays

# 2. Define the ripple function
# Calculate the radial distance from the center (0,0)
R = np.sqrt(X**2 + Y**2)

# Apply a damped sine wave function
# sin(5 * R) generates the wave pattern
# /(1 + R) dampens the waves as they move away from the center
Z = np.sin(5 * R) / (1 + R)

# 3. Plot the ripple effect
plt.figure(figsize=(6, 6))
plt.contourf(X, Y, Z, cmap='Blues') # Use a filled contour plot for visualization
plt.colorbar(label='Wave Intensity')
plt.axis('off') # Optional: remove axes for a cleaner look
plt.title("Water Ripple Pattern with Matplotlib")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Required for 3D plots

# 1. Create the coordinate grid and ripple function (same as above)
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(5 * R) / (1 + R)

# 2. Plot the ripple as a 3D surface
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='Blues', edgecolor='none')

ax.set_title("3D Water Ripple Surface with Matplotlib")
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (Height)')

# Optional: Rotate the view for a better angle
ax.view_init(elev=50., azim=-120)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Setup
total_degrees = np.linspace(0, 360, 50)
theta = np.radians(total_degrees)
radiuses = np.linspace(0, 10, 50)

# Initialize 2D matrices (50x50) instead of empty lists
# This provides the 2D structure plot_surface requires
x = np.zeros((len(theta), len(radiuses)))
y = np.zeros((len(theta), len(radiuses)))
z = np.zeros((len(theta), len(radiuses)))

# Fill matrices using nested loops
for i, t in enumerate(theta):
    for j, r in enumerate(radiuses):
        # Parametric equations for a circular ripple
        x[i, j] = r * np.cos(t)
        y[i, j] = r * np.sin(t)
        z[i, j] = np.sin(r)  # Height based on radius

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap="Blues")
plt.show()
```
